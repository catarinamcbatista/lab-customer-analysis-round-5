```python
# marketing_customer_analysis.csv
#x-y split
#normalize numerical


import pandas as pd

data = pd.read_csv("marketing_customer_analysis.csv")

data.head()
```

```python
data.shape
```

```python
#Total claim amount is going to be the target for this exercise
#x-y split: y is the variable for Total Claim Amount

y = data['Total Claim Amount']

x = data.drop('Total Claim Amount',axis=1)

x = x.select_dtypes('number')

x.head()
```

```python
display(y)
```

```python
#Normalize numerical data for x (dataframe with numerical values)

from sklearn.preprocessing import Normalizer

transformer = Normalizer().fit(x)

data_normalized = transformer.transform(x)

data_normalized
```
